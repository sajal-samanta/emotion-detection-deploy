```python
# CELL 1: First, install required packages if not already installed
import subprocess
import sys

def install_package(package):
    try:
        __import__(package)
        print(f"âœ“ {package} is already installed")
    except ImportError:
        print(f" Installing {package}...")
        
        subprocess.check_call([sys.executable, "-m", "pip", "install", package])

# List of required packages
required_packages = [
    "tensorflow",
    "opencv-python", 
    "pandas",
    "numpy",
    "matplotlib",
    "seaborn",
    "scikit-learn",
    "jupyter"
]

print("ðŸ”§ Checking and installing required packages...")
for package in required_packages:
    install_package(package)

print("\n" + "="*50)
print(" All packages installed successfully!")
print("="*50 + "\n")

# Now import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

# Import TensorFlow with error handling
try:
    import tensorflow as tf
    from tensorflow.keras.models import Sequential
    from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
    from tensorflow.keras.optimizers import Adam
    from tensorflow.keras.utils import to_categorical
    from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
    print(f"âœ“ TensorFlow version: {tf.__version__}")
except ImportError as e:
    print(f" TensorFlow import failed: {e}")
    print("Please install TensorFlow manually: pip install tensorflow")
    exit()

# Import other libraries
try:
    import cv2
    print(f"âœ“ OpenCV version: {cv2.__version__}")
except ImportError as e:
    print(f" OpenCV import failed: {e}")

import warnings
warnings.filterwarnings('ignore')

# Check GPU availability
if tf.test.gpu_device_name():
    print(f"âœ“ GPU is available: {tf.test.gpu_device_name()}")
else:
    print("  No GPU detected - running on CPU")

print("\n" + "="*50)
print(" All libraries imported successfully!")
print("="*50)

# Display versions for verification
print(f"\n Library Versions:")
print(f"Pandas: {pd.__version__}")
print(f"NumPy: {np.__version__}")
print(f"OpenCV: {cv2.__version__}")
print(f"TensorFlow: {tf.__version__}")
```

```python
# CELL:2 Load FER2013 dataset
data = pd.read_csv('../data/fer2013.csv')

print(f"Dataset shape: {data.shape}")
print("\nFirst 5 rows:")
print(data.head())

print("\nEmotion distribution:")
print(data['emotion'].value_counts())

# Emotion labels
emotion_labels = {
    0: 'Angry',
    1: 'Disgust',
    2: 'Fear',
    3: 'Happy',
    4: 'Sad',
    5: 'Surprise',
    6: 'Neutral'
}

plt.figure(figsize=(10, 6))
data['emotion'].value_counts().sort_index().plot(kind='bar')
plt.title('Emotion Distribution in Dataset')
plt.xlabel('Emotion')
plt.ylabel('Count')
plt.xticks(range(7), [emotion_labels[i] for i in range(7)], rotation=45)
plt.show()
```

```python
#CELL3: 
def preprocess_data(data):
    # Convert pixels to numpy arrays
    pixels = data['pixels'].apply(lambda x: np.array(x.split(), dtype='float32'))
    
    # Reshape to 48x48x1
    X = np.stack(pixels.values, axis=0).reshape(-1, 48, 48, 1)
    
    # Normalize pixel values
    X = X / 255.0
    
    # Convert emotions to categorical
    y = to_categorical(data['emotion'].values, num_classes=7)
    
    return X, y

# Split data
train_data = data[data['Usage'] == 'Training']
test_data = data[data['Usage'] == 'PublicTest']
val_data = data[data['Usage'] == 'PrivateTest']

print(f"Training samples: {len(train_data)}")
print(f"Test samples: {len(test_data)}")
print(f"Validation samples: {len(val_data)}")

# Preprocess all datasets
X_train, y_train = preprocess_data(train_data)
X_test, y_test = preprocess_data(test_data)
X_val, y_val = preprocess_data(val_data)

print(f"X_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")
```

```python
#CELL 4
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Create data generator for augmentation
datagen = ImageDataGenerator(
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    zoom_range=0.1,
    shear_range=0.1
)

datagen.fit(X_train)

# Visualize augmented images
plt.figure(figsize=(12, 6))
for i in range(8):
    plt.subplot(2, 4, i+1)
    augmented = datagen.flow(X_train[:8], y_train[:8], batch_size=1)[0][0]
    plt.imshow(augmented.reshape(48, 48), cmap='gray')
    plt.axis('off')
plt.suptitle('Augmented Training Images')
plt.show()
```

```python
#CELL 5
def create_emotion_model():
    model = Sequential([
        # First Convolutional Block
        Conv2D(64, (3, 3), activation='relu', padding='same', input_shape=(48, 48, 1)),
        BatchNormalization(),
        Conv2D(64, (3, 3), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling2D(2, 2),
        Dropout(0.25),
        
        # Second Convolutional Block
        Conv2D(128, (3, 3), activation='relu', padding='same'),
        BatchNormalization(),
        Conv2D(128, (3, 3), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling2D(2, 2),
        Dropout(0.25),
        
        # Third Convolutional Block
        Conv2D(256, (3, 3), activation='relu', padding='same'),
        BatchNormalization(),
        Conv2D(256, (3, 3), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling2D(2, 2),
        Dropout(0.25),
        
        # Fourth Convolutional Block
        Conv2D(512, (3, 3), activation='relu', padding='same'),
        BatchNormalization(),
        Conv2D(512, (3, 3), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling2D(2, 2),
        Dropout(0.25),
        
        # Fully Connected Layers
        Flatten(),
        Dense(1024, activation='relu'),
        BatchNormalization(),
        Dropout(0.5),
        Dense(512, activation='relu'),
        BatchNormalization(),
        Dropout(0.5),
        Dense(7, activation='softmax')
    ])
    
    return model

model = create_emotion_model()
model.summary()
```

```python
# Add this cell BEFORE training for better accuracy
from tensorflow.keras import regularizers

# Enhanced data augmentation for better generalization
enhanced_datagen = ImageDataGenerator(
    rotation_range=20,  # Increased from 15
    width_shift_range=0.15,  # Increased
    height_shift_range=0.15,  # Increased
    horizontal_flip=True,
    zoom_range=0.2,  # Increased
    shear_range=0.15,  # Increased
    fill_mode='nearest'
)

enhanced_datagen.fit(X_train)
```

```python
# Cell 6: Fixed Training Code

print("âš¡ Configuring TensorFlow for optimal performance...")

# Enable XLA compilation
tf.config.optimizer.set_jit(True)

# Set memory growth for GPU
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        print("âœ“ GPU memory growth enabled")
    except RuntimeError as e:
        print(e)

# REMOVE mixed precision for now - it's causing issues
# try:
#     policy = tf.keras.mixed_precision.Policy('mixed_float16')
#     tf.keras.mixed_precision.set_global_policy(policy)
#     print("âœ“ Mixed precision enabled")
# except:
#     print("â„¹ï¸ Mixed precision not available, using FP32")

print("âœ“ Performance optimizations applied!\n")

# Use this OPTIMIZED model architecture instead
def create_optimized_emotion_model():
    model = Sequential([
        # First Convolutional Block
        Conv2D(64, (3, 3), activation='relu', padding='same', input_shape=(48, 48, 1)),
        BatchNormalization(),
        Conv2D(64, (3, 3), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling2D(2, 2),
        Dropout(0.25),
        
        # Second Convolutional Block
        Conv2D(128, (3, 3), activation='relu', padding='same'),
        BatchNormalization(),
        Conv2D(128, (3, 3), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling2D(2, 2),
        Dropout(0.25),
        
        # Third Convolutional Block
        Conv2D(256, (3, 3), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling2D(2, 2),
        Dropout(0.25),
        
        # Fourth Convolutional Block
        Conv2D(512, (3, 3), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling2D(2, 2),
        Dropout(0.25),
        
        # Fully Connected Layers
        Flatten(),
        Dense(512, activation='relu'),
        BatchNormalization(),
        Dropout(0.5),
        Dense(256, activation='relu'),
        Dropout(0.5),
        Dense(7, activation='softmax')
    ])
    
    return model

# Create the optimized model
print("ðŸ”„ Creating optimized model...")
model = create_optimized_emotion_model()

# Enhanced Callbacks
reduce_lr = ReduceLROnPlateau(
    monitor='val_accuracy',
    factor=0.2,
    patience=3,
    min_lr=0.0001,
    verbose=1
)

early_stop = EarlyStopping(
    monitor='val_accuracy',
    patience=8,
    restore_best_weights=True,
    verbose=1,
    min_delta=0.001
)

# Compile model
print("ðŸ”„ Compiling model...")
model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("âœ… Model compiled successfully!")
print(f"ðŸ“Š Training on {len(X_train)} samples")
print(f"ðŸ“Š Validating on {len(X_val)} samples")
print(f"ðŸŽ¯ Batch size: 128")
print(f"â° Epochs: 50 (will stop early if converged)")

# Train model
print("\nðŸš€ Starting optimized training...")
history = model.fit(
    datagen.flow(X_train, y_train, batch_size=128),
    epochs=50,
    validation_data=(X_val, y_val),
    callbacks=[reduce_lr, early_stop],
    verbose=1
)

print("ðŸŽ‰ Training completed!")
```

```python
# Evaluate model
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_accuracy:.4f}")
print(f"Test Loss: {test_loss:.4f}")

# Plot training history
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

ax1.plot(history.history['accuracy'], label='Training Accuracy')
ax1.plot(history.history['val_accuracy'], label='Validation Accuracy')
ax1.set_title('Model Accuracy')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Accuracy')
ax1.legend()

ax2.plot(history.history['loss'], label='Training Loss')
ax2.plot(history.history['val_loss'], label='Validation Loss')
ax2.set_title('Model Loss')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Loss')
ax2.legend()

plt.show()
```

```python
# Predictions
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(y_test, axis=1)

# Classification Report
print("Classification Report:")
print(classification_report(y_true_classes, y_pred_classes, 
                          target_names=[emotion_labels[i] for i in range(7)]))

# Confusion Matrix
plt.figure(figsize=(10, 8))
cm = confusion_matrix(y_true_classes, y_pred_classes)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=[emotion_labels[i] for i in range(7)],
            yticklabels=[emotion_labels[i] for i in range(7)])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```

```python
# Cell 9: Enhanced Model Saving with Better Practices

print("ðŸ’¾ Saving trained model...")

# Create models directory if it doesn't exist
import os
os.makedirs('models', exist_ok=True)

# Option 1: Save as modern Keras format (recommended - no warnings)
model.save('models/emotion_model.keras')
print("âœ… Model saved in Keras format: 'models/emotion_model.keras'")

# Option 2: Also save as HDF5 for compatibility (will show warning but works)
model.save('models/emotion_model.h5')
print("âœ… Model saved in HDF5 format: 'models/emotion_model.h5'")

# Option 3: Save model architecture and weights separately
# Save architecture as JSON
model_json = model.to_json()
with open("models/emotion_model_architecture.json", "w") as json_file:
    json_file.write(model_json)
print("âœ… Model architecture saved: 'models/emotion_model_architecture.json'")

# Save weights with correct extension (.weights.h5)
model.save_weights("models/emotion_model.weights.h5")  # Fixed extension
print("âœ… Model weights saved: 'models/emotion_model.weights.h5'")

# Save training history for future analysis
import pickle
with open('models/training_history.pkl', 'wb') as f:
    pickle.dump(history.history, f)
print("âœ… Training history saved: 'models/training_history.pkl'")

# Save TensorFlow Lite model for mobile/edge deployment
print("\nðŸ“± Converting to TensorFlow Lite...")
try:
    converter = tf.lite.TFLiteConverter.from_keras_model(model)
    
    # Enable optimizations for smaller size and faster inference
    converter.optimizations = [tf.lite.Optimize.DEFAULT]
    
    tflite_model = converter.convert()
    
    # Save TensorFlow Lite model
    with open('models/emotion_model.tflite', 'wb') as f:
        f.write(tflite_model)
    
    # Print model size information
    tflite_size = len(tflite_model) / (1024 * 1024)  # Convert to MB
    print(f"âœ… TensorFlow Lite model saved: 'models/emotion_model.tflite'")
    print(f"ðŸ“¦ Model size: {tflite_size:.2f} MB")
    
except Exception as e:
    print(f"âŒ Error converting to TensorFlow Lite: {e}")

# Model performance summary
print("\n" + "="*50)
print("ðŸŽ¯ TRAINING SUMMARY")
print("="*50)

# Get final metrics
final_train_acc = history.history['accuracy'][-1]
final_val_acc = history.history['val_accuracy'][-1]
final_train_loss = history.history['loss'][-1]
final_val_loss = history.history['val_loss'][-1]

print(f"ðŸ“Š Final Training Accuracy: {final_train_acc:.4f} ({final_train_acc*100:.2f}%)")
print(f"ðŸ“Š Final Validation Accuracy: {final_val_acc:.4f} ({final_val_acc*100:.2f}%)")
print(f"ðŸ“‰ Final Training Loss: {final_train_loss:.4f}")
print(f"ðŸ“‰ Final Validation Loss: {final_val_loss:.4f}")

# Find best validation accuracy
best_val_acc = max(history.history['val_accuracy'])
best_epoch = history.history['val_accuracy'].index(best_val_acc) + 1
print(f"ðŸ† Best Validation Accuracy: {best_val_acc:.4f} ({best_val_acc*100:.2f}%) at Epoch {best_epoch}")

print("\nðŸ’¾ All model files saved successfully in 'models/' directory!")
print("ðŸŽ‰ Ready for deployment in the Streamlit web application!")
```
